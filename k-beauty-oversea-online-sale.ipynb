# 데이터 로드하기

```python
df_raw = pd.read_csv("C:/data/국가_대륙_별_상품군별_온라인쇼핑_해외직접판매액1.csv", encoding="cp949")
df_raw.shape
```

# 분석과 시각화를 위한 tidy data 만들기

```python
df_raw.columns
```

```python
# column에 있는 데이터를 행으로
# id_vars : id값으로 지정하면 나머지 column값들은 자동으로 value로 지정됨
df = df_raw.melt(id_vars = ["국가(대륙)별", "상품군별", "판매유형별"], var_name = "기간", value_name = "백만원")
df.shape
```

```python
df.head()
```

```python
df.info() # object : String 데이터를 의미
```

# 데이터 전처리

## 기간에서 연도를 분리하기

```python
int("2019 4/4".split()[0]) # 연도 데이터 분리 후 정수형으로 타입 변경
```

```python
# map : 함수를 구현하여 적용시킬 수 있음
df["연도"] = df["기간"].map(lambda x : int(x.split()[0])) # 연도 데이터만 분리하고 타입 변경하여 '연도' 컬럼 생성
df.head()
```

```python
"2019 4/4".split()[1].split("/")[0] # 분기 데이터로 분리
```

```python
# 기간에서 분기만 분리하기
df["분기"] = df["기간"].map(lambda x : int(x.split()[1].split("/")[0])) # '분기' 컬럼 생성
df.head()
```

## 금액을 수치데이터로 표현하기 위해 데이터 타입 변경하기

```python
# - 문자를 결측치로 변경하고 float 타입으로 변경하기 (NaN이라는 결측치가 float 타입이기 때문)
df["백만원"] = df["백만원"].replace("-", pd.np.nan).astype(float) # - 문자를 결측치로 변경
df.head()
```

## 필요없는 데이터 제거하기

```python
df.info() # 메모리 용량 보기
```

```python
# 합계 데이터는 따로 구할 수 있기 때문에 전체 데이터에서 제거
df = df[(df["국가(대륙)별"] != "합계") & (df["상품군별"] != "합계")].copy()
df.info() # 메모리 용량이 훨씬 줄어든 것을 볼 수 있음
```

```python
# 결측치 보기
df.isnull().sum() # true값만 더하기 -> 금액이 없는 데이터도 있음
```

# K-Beauty 시각화

## 전체 상품군 판매액

```python
# 판매유형별 데이터는 일부 기간에는 "계"만 존재하기 때문에 
df_total = df[df["판매유형별"] == "계"].copy()
df_total.head() # 상품군별로 데이터가 있는 것을 확인할 수 있음
```

```python
# 연도, 판매액 lineplot으로 그리기
sns.lineplot(data = df_total, x = "연도", y = "백만원") # 판매액이 계속 증가하고 있음을 확인
```

```python
# 연도, 판매액 lineplot으로 그리고 상품군별로 다른 색상으로 표시하기
sns.lineplot(data = df_total, x = "연도", y = "백만원", hue = "상품군별")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) # legend를 그래프의 밖에 그리기
# 화장품이 굉장히 증가했다는 것을 확인할 수 있음
```

```python
# 위에 그린 그래프를 자세히 보기 위해 서브플롯으로 표시하기
# col_warp = n : 한 줄에 n개 표현
sns.relplot(data = df_total, x = "연도", y = "백만원", hue = "상품군별", kind = "line", col = "상품군별", col_wrap = 4)
# 화장품값이 너무 커서 다른 값을 보기가 어려움
```

```python
# isin을 사용해 화장품만 제외를 하고 df_sub이라는 변수에 담기
df_sub = df_total[~df_total["상품군별"].isin(["화장품", "의류 및 패션 관련상품"])].copy()
df_sub
```

```python
# 연도별 판매액을 상품군별로 relplot을 활용해 서브블롯으로 그리기
sns.relplot(data = df_sub, x = "연도", y = "백만원", hue = "상품군별", col = "상품군별", col_wrap = 4, kind = "line")
```

## 화장품의 온라인쇼핑 해외직접판매액

```python
# df_cosmetic 이라는 변수에 상품군별이 화장품인 데이터만 가져오기
df_cosmetic = df_total[df_total["상품군별"] == "화장품"].copy()
df_cosmetic["상품군별"].unique() # 화장품만 들어있음
```

```python
# 연도와 판매액을 lineplot으로 그리고 분기별로 다른 색상으로 표현해보기
plt.figure(figsize=(15, 4))
sns.lineplot(data = df_cosmetic, x = "연도", y = "백만원", hue = "분기")
# 꾸준히 증가하고 있기 때문에 4분기가 판매량이 좋다는 것을 알 수 있음
```

```python
df_cosmetic.head()
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화 하기
plt.figure(figsize=(15, 4))
plt.xticks(rotation=30) # 30도 정도 그래프 글씨 기울기
sns.lineplot(data=df_cosmetic, x="기간", y="백만원")
# 화장품 판매액은 꾸준히 증가해옴
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화하고 "국가(대륙)별"로 다른 색상으로 표시하기
plt.figure(figsize=(15, 4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic, x="기간", y="백만원", hue = "국가(대륙)별")
```

```python
# "중국" 제외하고 보기
plt.figure(figsize=(15, 4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic[df_cosmetic["국가(대륙)별"] != "중국"], x="기간", y="백만원", hue = "국가(대륙)별")
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화하고 "판매유형별"로 다른 색상으로 표시하기
# 2016년 이전에는 데이터가 존재하지 않음 -> 2017년부터 그릴 수 있음
plt.figure(figsize=(15, 4))
plt.xticks(rotation=30)
# df_sub = df[df["판매유형별"] != "계"].copy()
df_sub = df[(df["판매유형별"] != "계") & (df["판매유형별"] != "면세점")].copy() # 면세점 제외
sns.lineplot(data=df_sub, x="기간", y="백만원", hue="판매유형별", ci=None)
# 1. 온라인 판매액에 꾸준히 증가하고 있음
# 2. 전체적으로 성장하고 있음
```

## 의류 및 패션관련 상품 온라인쇼핑 해외직접판매액

```python
# df_fashion 이라는 변수에 의류 데이터만 가져와 따로 담아주기
# "계" 데이터만 따로 가져와서 보기 (나머지 결측치 때문에)
df_fashion = df[(df["상품군별"] == "의류 및 패션 관련상품") & (df["판매유형별"] == "계")].copy()
df_fashion.head()
```

```python
# "의류 및 패션 관련상품" 판매액에 대한 기간별 금액 데이터 시각화하고 "국가(대륙)별"로 다른 색상으로 표시하기
plt.figure(figsize=(15, 4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion, x="기간", y="백만원", hue="국가(대륙)별")
```

```python
# "의류 및 패션 관련상품" 판매액에 대한 기간별 금액 데이터 시각화하고 "판매유형별"로 다른 색상으로 표시하기
df_fashion2 = df[(df["상품군별"] == "의류 및 패션 관련상품") & (df["판매유형별"] != "계")].copy() # "계"만 제거하고 가져오기

plt.figure(figsize=(15, 4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion2, x="기간", y="백만원", hue="판매유형별", ci=None)
```

## 데이터 집계하기

```python
df_fashion.head(2)
```

```python
# 피봇테이블로 "국가(대륙)별", "연도"별 합계 금액을 표 형태로 구하기
result= df_fashion.pivot_table(index="국가(대륙)별", columns="연도", values="백만원", aggfunc="sum")
result
```

## 연산결과를 시각적으로 보기

```python
# 피봇테이블로 구한 결과를 값의 많고 적음에 따라 시각적으로 표현하기
# vmin, vmax : 수치값 설정
plt.figure(figsize=(10, 6))
sns.heatmap(result, cmap="Blues", annot=True, fmt=".0f") # 많을수록 색이 진해짐
# 적을수록 색이 진해지는 경우 -> Blues_r
# annot : 숫자값 표시, fmt : 소수점 없이 float 형태로 표시
```
